### 2. Read the first n numbers from ages.txt into a list ages. In case the file


````python
n = int(input("Enter n: "))
print(f"n = {n}")
with open("ages.txt", "r") as f:
    content = f.read()
    allData = [float(x) for x in content.split(",")]
    if n > len(allData):
        print("elements in file less than n")
    else:
        ages = allData[:n]
        print(ages)

````

### 3. Round the numbers to the lowest integer and modify the list ages.


````python
import math
ages = [math.floor(x) for x in ages]
print(ages)
````

### 4. Create a file scores.csv with the following data:

```csv
Name,Score
Andrew,88.3
Ben,92.6
Carol,89.7
```


### 5. Read data from scores.csv into a dictionary dic.


````python
people = {'name': [], 'score': []}
with open('scores.csv', 'r') as f:
    lines = f.readlines()[1:]   # skip first line
    for line in lines:
        [name, score] = line.strip().split(',')
        people['name'].append(name)
        people['score'].append(float(score))

print(people)

    
````

### 6. Add the list ages into dic as a new column after the column Name.


````python
people['age'] = []
for i in range(min(len(ages), len(people['name']))):
    people['age'].append(ages[i])
# order keys in dictionary 
people  = {
    'name': people['name'],
    'age': people['age'],
    'score': people['score'],
    }
print(people) 
````

### 7. Convert the dictionary into a data frame df. Print the data frame.


````python
import pandas as pd
df = pd.DataFrame(people)
print(df)
````

### 8. Select the column ages and find the maximum value. Print the result.

````python
column = 'Daniel,10,78'
[name, age, score] = column.split(',')
df.loc[len(df)] = [name, int(age), float(score)]
print(df)
````

### 10. Modify the Score value in the second row (Ben,15,92.6) by applying a deduction p


````python
p = float(input("Enter deduction p: "))
df.loc[1, 'score'] = df.loc[1, 'score'] - p
print(df)
````

### 11.Round the values in the Score column to the upper half-point.

```txt
For example, 2.3 is rounded to 2.5
 2.6 is rounded to 3
 2 is rounded to 2
 2.5 is rounded to 2.5.
```


````python
import numpy as np
df['score'] = np.ceil(df["score"] * 2) / 2
print(df)
````

### 12.Find the mean and median scores in df. Print the updated data frame, mean and median scores.


````python
mean = df['score'].mean()
median = df['score'].median()
print(f"Mean: {mean}, Median: {median}")
````

### 13.Create a new data frame `df_filtered` containing only rows with ages between 12 and 15, inclusive. Print `df_filtered`.


````python
df_filtered = df[(df['age'] >= 12) & (df['age'] <= 15)]
print(df_filtered)
````
