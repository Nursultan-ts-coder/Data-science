3. Construct a pie chart showing the percentages of Oscar winners by gender


```python
import pandas as pd
import matplotlib.pyplot as plt

import pandas as pd
df = pd.read_excel("oscars.xlsx", engine="openpyxl")
df = pd.read_excel("oscars.xlsx")
maleCount = df[(df['gender'] == 'Male')].shape[0]
femaleCount = df[(df['gender'] == 'Female')].shape[0]

sizes = [maleCount, femaleCount]
labels =  ['Male', 'Female']
colors = ["#009ab9", "#ff7c18"]

plt.pie(
    x=[maleCount, femaleCount],
    labels=['Male', 'Female'],
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops={'edgecolor': 'white', 'linewidth': 1},
    colors=colors
)

plt.title("Oscar winners by gender", pad=12)
plt.axis('equal')

plt.savefig("gender_distribution.png", bbox_inches="tight")
plt.show()
```

5. Construct a bar chart showing the number of winners across all categories.
   Save the chart as `winners_by_category.png` to your working
   directory.


```python
import matplotlib.pyplot as plt
categories = df['Category'].value_counts()

values = categories.values
labels = categories.index
labels = [label.capitalize() for label in labels]

plt.figure(figsize=(5, 4), dpi=100)
plt.barh(labels, values)
plt.title("Oscar winners by category")
plt.xlabel("Count")
plt.ylabel("Category")
plt.show()
```

7. Construct a bar chart showing the number of winners for the n most frequent
   categories. Save the chart as `top_n_winners_by_category`.png
   with the specified value of n.


```python
import matplotlib.pyplot as plt
top_n = input("Enter the number of top categories to display: ")
top_n = int(top_n)
top = df['Category'].value_counts().head(top_n)

values = top.values
labels = top.index
labels = [label.capitalize() for label in labels]

plt.figure(figsize=(5, 4), dpi=100)
plt.barh(labels, values)
plt.title(f"Top {top_n} Oscar winners by category")
plt.xlabel("Count")
plt.ylabel("Category")
plt.savefig(f"top_{top_n}_categories.png", bbox_inches="tight")
plt.show()
```
